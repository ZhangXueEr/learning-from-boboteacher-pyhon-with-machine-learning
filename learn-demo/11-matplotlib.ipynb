# matplotlib基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red')
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='-.') # linestyle: ":" "-." "--" "-"
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='-.') # linestyle: ":" "-." "--" "-"
plt.xlim(-5, 15)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='-.') # linestyle: ":" "-." "--" "-"
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='-.') # linestyle: ":" "-." "--" "-"
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='-.') # linestyle: ":" "-." "--" "-"
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, color='red', linestyle='-.', label='cos(x)') # linestyle: ":" "-." "--" "-"
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, color='red', linestyle='-.', label='cos(x)') # linestyle: ":" "-." "--" "-"
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("welcom to the ML world!")
plt.show()
```

## Scatter Plot

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color='red')
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y, alpha=0.5) # alpha不透明度，0完全透明，1不透明
plt.show()
```

```python
plt.scatter(x, y, alpha=0.1)
plt.show()
```
